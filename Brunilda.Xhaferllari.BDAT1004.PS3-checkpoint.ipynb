```python
# Step 1. Import the necessary libraries
import pandas as pd
import datetime as dt
```

```python
# Step 2. Import the dataset from this address
windData = pd.read_csv(r'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data',sep = '\s+')
windData.head()
```

```python
windData.columns
```

```python
#Step 3. Assign it to a variable called data and replace the first 3 columns by a proper datetime index.
data = ['Dy','Mo','Yr']
windData['FullDate'] = windData[data].apply(lambda row: '-'.join(row.values.astype(str)), axis = 'columns') #create a new column
windData = windData.drop(columns=['Yr','Mo','Dy'])    #delete the 3 columns
windData                          
```

```python
#windData.to_datetime('FullDate', format='%Y%m%d')
windData['FullDate'] = pd.to_datetime(windData.FullDate) #changed the format in date
windData
```

```python
windData.FullDate.dtype
```

```python
#Step 4. Year 2061? Do we really have data from this year? Create a function to fix it and apply it.
def correctdate(x):
    if x.year > 2021:
        year = x.year - 100
    else:
        year = x.year
    return datetime.date(year,x.month,x.day)

windData['FullDate'] = windData['FullDate'].apply(correctdate)  # Changed the year considering that all years greater than 2021.
windData.sort_values(by=['FullDate'])#orderd the table we se that the years are from 78-61 that seam normal
windData['FullDate'] = pd.to_datetime(windData.FullDate)
windData.head()
```

```python
windData.FullDate.dtype
```

```python
#Step 5. Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns].
windData.set_index('FullDate',inplace=True)
windData
```

```python
windData.info()
```

```python
#Step 6. Compute how many values are missing for each location over the entire record.They should be ignored in all calculations below.
windDataNan=windData.isnull().values.ravel().sum() #https://numpy.org/doc/stable/reference/generated/numpy.ravel.html
windDataNan
```

```python
#Step 7. Compute how many non-missing values there are in total.
windDataNanN=windData.notnull().values.ravel().sum()
windDataNanN
```

```python
#Step 8. Calculate the mean windspeeds of the windspeeds over all the locations and all the times.A single number for the entire dataset.
meanColumn=windData.mean() #the mean of each location
meanColumn.mean()      #the mean of each column
```

```python
#Step 9. Create a DataFrame called loc_stats and calculate the min, max and mean windspeeds and standard deviations of the windspeeds at each location over all the days
#A different set of numbers for each location.
loc_stats=windData.agg(['min','max','mean','std']) #https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.agg.html?highlight=agg
loc_stats
```

```python
#Step 10. Create a DataFrame called day_stats and calculate the min, max and mean windspeed and standard deviations of the windspeeds across all the locations at each day.
day_stats=windData.apply(pd.DataFrame.describe, axis=1)
day_stats
```

```python
#Step 11. Find the average windspeed in January for each location.
windspeedM = windData.groupby(lambda d: d.month==1) #group by month
monthly_means = windspeedM.mean()                   #mean for each location
monthly_means
```

```python
#Step 12. Downsample the record to a yearly frequency for each location.
windData.groupby(windData.index.to_period('A')).mean()
```

```python
#Step 13. Downsample the record to a monthly frequency for each location.
windData.groupby(windData.index.to_period('M')).mean()
```

```python
#Step 14. Downsample the record to a weekly frequency for each location.
windData.groupby(windData.index.to_period('W')).mean()
```
